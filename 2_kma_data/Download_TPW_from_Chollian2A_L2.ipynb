## 2. Python Library와 자료취득 기간 설정

```python
import pandas
from datetime import datetime
import wget

start = datetime(2022, 2, 4, 11, 10)
end = datetime(2022, 3, 31, 23, 50)
```

## 3. 자료취득기간내 10분단위 시간자료 생성

```python
dates = pandas.date_range(start=start, end=end, freq='10T')
dates
```

## 4. 천리안 자료 명칭지정을 위해 Date 포맷을 Text로 변환

```python
data_period = []
for date in dates:
    date_time = datetime.strptime(str(date), '%Y-%m-%d %H:%M:%S')
    date_time_str = date_time.strftime('%Y%m%d%H%M')
    data_period.append(date_time_str)
data_period
```

## 5. 천리안자료 download 시작

 * 저장위치 변경시 wget명령 뒤의 'TPW/TPW_{value}.nc'자료 변경필요

```python
for value in data_period:
    try:
        !wget "http://api.nmsc.kma.go.kr:9080/api/GK2A/LE2/TPW/KO/data?date={value}&key=NMSCe084741aa1354e068d1715e630656ce2" -l=1 -O TPW/TPW_{value}.nc
    except:
        pass
```
